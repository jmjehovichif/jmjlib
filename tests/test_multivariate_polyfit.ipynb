```python
import numpy as np
import matplotlib.pyplot as plt

import sys
import os

sys.path.append(os.path.abspath('..'))

from jmjlib.multivariate_polyfit import MultivariatePolynomialFitting
```

```python
# True function: y = 0.5*x^2 - 2*x + 3 + noise
np.random.seed(42)
X = np.linspace(-5, 5, 50).reshape(-1, 1)
Y = 0.5*X**2 - 2*X + 3 + np.random.normal(0, 2, size=X.shape)

plt.scatter(X, Y, color="blue", label="data")
plt.legend()
plt.show()
```

```python
model = MultivariatePolynomialFitting(degree=2)
model.set_data(X, Y)
model.fit()

print("R^2 Score:", model.score())
```

```python
X_test = np.linspace(-6, 6, 100).reshape(-1, 1)
Y_pred = model.predict(X_test)

plt.scatter(X, Y, color="blue", label="data")
plt.plot(X_test, Y_pred, color="red", label="fit")
plt.legend()
plt.show()
```
